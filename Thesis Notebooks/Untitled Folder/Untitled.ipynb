```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from fbm import FBM
%matplotlib inline  
```

```python
tau = 365
T = tau
H = 0.6
f = FBM(n=tau, hurst=H, length=T, method='daviesharte')
fbm_sample = pd.Series(f.fbm())

z = fbm_sample.diff()

#z.hist(bins = 40);
print('mean: ' + str(z.mean()))
print('std dev: ' + str(z.std()))
plt.figure()
plt.plot(z.cumsum());
```

```python
T = 1
N = 10000
H = [0.01,0.25,0.37,0.5,0.67,0.75,0.9]
z = []

for i in H:
    
    f = FBM(n=N, hurst=i, length=T, method='daviesharte')
    fbm_sample = pd.Series(f.fbm())
    
    z.append(fbm_sample.diff().std())
plt.plot(H,z);

plt.plot(f.fbm())
```
